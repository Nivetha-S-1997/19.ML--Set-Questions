```python
#8. Answer:
import numpy as np
import pandas as pd
from sklearn.linear_model import LinearRegression
from sklearn.impute import SimpleImputer

# Creating a dataset with missing values
X_train = np.array([[1, 2], [3, np.nan], [5, 6]])
y_train = np.array([10, 20, 30])

simp_imp=SimpleImputer(strategy='mean')
X_train_pre=simp_imp.fit_transform(X_train)

model = LinearRegression()
model.fit(X_train_pre, y_train)
```

```python
#10. Answer:
from sklearn.svm import SVC
from sklearn.preprocessing import StandardScaler

X_train = [[1, 100], [2, 200], [3, 300]]
y_train = [0, 1, 0]

scaler=StandardScaler()
X_train_scaler=scaler.fit_transform(X_train)

model = SVC()
model.fit(X_train_scaler, y_train)
predict_scaler=scaler.transform([[1, 150]])
print(model.predict(predict_scaler))
```
